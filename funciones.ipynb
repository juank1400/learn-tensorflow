# Activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Defining sigmoid function

```python
def sigmoid(n):
    return 1/(1+np.exp(-n))
```

### Plotting sigmoid function

```python
x = np.linspace(-20,20,400)
plt.plot(x,sigmoid(x))
plt.show()
```

## Defining scaloned function

```python
def scaloned(m):
    """

    Args:
      m: variable to be evaluated
    Returns:
      1 if m>0, 0 otherwise
    """
    if m>0:
        return 1
    else:
        return 0
```

### Plotting scaloned function

```python
x = np.linspace(-20,20,400)
results = [scaloned(i) for i in x]
plt.plot(x,results)
plt.show()
```

## Defining function relu

```python
def relu(o):
    if o>0:
        return o
    else:
        return 0
```

### Plotting relu function

```python
# Plotting relu function
x = np.linspace(-20,20,400)
results = [relu(i) for i in x]
plt.plot(x,results)
plt.show()
```

## Defining tanh function

```python
def tanh(o):
    return np.tanh(o)
```

### Plotting tanh function

```python
# Plotting relu function
x = np.linspace(-5,5,400)
results = [tanh(i) for i in x]
plt.plot(x,results)
plt.show()
```

# Loss functions

## MSE

### Defining Loss functions

```python
def MSE(y,y_hat):
    return np.mean((y-y_hat)**2)

def MAE(y,y_hat):
    return np.mean(np.abs(y-y_hat))

def RMSE(y,y_hat):
    return np.sqrt(np.mean((y-y_hat)**2))

def R2(y,y_hat):
    return 1-(np.sum((y-y_hat)**2)/np.sum((y-np.mean(y))**2))

def cross_entropy(y,y_hat):
    return -np.mean(y*np.log(y_hat) + (1-y)*np.log(1-y_hat))
```

### Plotting MSE function

```python
# Plotting relu function
results = np.linspace(0,20,400)
results_predicted = np.random.random(400)*20

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Graficar el primer arreglo
plt.plot(results, label='Predicted', linestyle='-', marker='o')

# Graficar el segundo arreglo
plt.plot(results_predicted, label='expected', linestyle='-', marker='x')

# Aadir ttulo y etiquetas
plt.title('Comparacin de Dos Arreglos')
plt.xlabel('ndice')
plt.ylabel('Valores')

# Mostrar la leyenda
plt.legend()

# Mostrar la grfica
plt.grid(True)
plt.show()
```

```python
loss = [ ["MSE", MSE(results,results_predicted)], ["MAE", MAE(results,results_predicted)], ["RMSE", RMSE(results,results_predicted)], ["R2", R2(results,results_predicted)]] 
loss
```
